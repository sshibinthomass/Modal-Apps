# Deploy and Run Llama-3.2-3B-Instruct on Modal

This notebook deploys the model to Modal and runs it using the deployed class.

```python
import modal

LlamaModel = modal.Cls.from_name("llama-3-2-3b-instruct", "LlamaModel")
model = LlamaModel()

prompt = "Tell me a short joke about programming."
reply = model.generate.remote(prompt)
print(reply)
```

```python
# Test with a structured chat template (messages)
messages = [
    {"role": "system", "content": "You are a precise assistant."},
    {"role": "user", "content": "What are the three laws of thermodynamics?"}
]
reply = model.generate.remote(messages)
print(reply)
```

# Deploy and Run Gemma 4 12B IT on Modal

This notebook deploys the Gemma-4-12b-it model to Modal and runs inference using the deployed class.

```python
# Deploy the Gemma model to Modal
!uv run modal deploy ../llm-hosting/gemma-4-12B-it.py
```

```python
import modal

GemmaModel = modal.Cls.from_name("gemma-4-12b-it", "GemmaModel")
model = GemmaModel()

prompt = "Explain the difference between supervised and unsupervised learning in 2 sentences."
reply = model.generate.remote(prompt)
print(reply)
```

```python
# Test with a structured chat template (messages)
messages = [
    {"role": "system", "content": "You are a precise assistant."},
    {"role": "user", "content": "What are the three laws of thermodynamics?"}
]
reply = model.generate.remote(messages)
print(reply)
```

```python
# Test with thinking mode enabled (enable_thinking=True)
messages = [
    {"role": "system", "content": "You are a helpful assistant."},
    {"role": "user", "content": "How many Rs are in the word strawberry?"}
]
reply = model.generate.remote(messages, enable_thinking=True)
print(reply)
```

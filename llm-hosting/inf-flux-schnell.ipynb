# Deploy and Run FLUX.1-schnell on Modal

This notebook deploys the FLUX.1-schnell text-to-image model to Modal and runs image generation inference using an L4 GPU.

```python
# Deploy the FLUX model to Modal
!uv run modal deploy flux-schnell.py
```

```python
import modal
from PIL import Image
import io

# Load the deployed FluxModel
FluxModel = modal.Cls.from_name("flux-1-schnell", "FluxModel")
flux_model = FluxModel()

prompt = "A beautiful, cinematic image of a futuristic city with flying cars at sunset, highly detailed"
print(f"Generating image for prompt: {prompt}")
image_bytes = flux_model.generate.remote(prompt)

# Display the generated image in the notebook
image = Image.open(io.BytesIO(image_bytes))
display(image)
```
